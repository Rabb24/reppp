```python
import pandas as pd

def count_nulls_by_category(csv_file, category_column):
    # Wczytaj dane z pliku CSV
    df = pd.read_csv(csv_file)

    # Inicjalizacja słownika do przechowywania wyników
    null_counts_dict = {}

    # Iteracja po kolumnach
    for column in df.columns:
        if column != category_column:
            # Liczenie nulli w kolumnie dla kategorii 0
            nulls_for_0 = df[df[category_column] == 0][column].isnull().sum()
            # Liczenie nulli w kolumnie dla kategorii 1
            nulls_for_1 = df[df[category_column] == 1][column].isnull().sum()

            # Dodanie wyników do słownika
            null_counts_dict[column] = {'nulls_for_0': nulls_for_0, 'nulls_for_1': nulls_for_1}

    return null_counts_dict

# Przykład użycia funkcji
csv_file = 'twoj_plik.csv'  # Zamień na nazwę swojego pliku CSV
category_column = 'wybrana_kolumna'  # Zamień na nazwę kolumny, której wartości są 0 lub 1

null_counts = count_nulls_by_category(csv_file, category_column)
print(null_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def count_nulls_by_category(csv_file, category_column):
    # Wczytaj dane z pliku CSV
    df = pd.read_csv(csv_file)

    # Inicjalizacja słownika do przechowywania wyników
    null_counts_dict = {}

    # Iteracja po kolumnach
    for column in df.columns:
        if column != category_column:
            # Liczenie nulli w kolumnie dla kategorii 0
            nulls_for_0 = df[df[category_column] == 0][column].isnull().sum()
            # Liczenie nulli w kolumnie dla kategorii 1
            nulls_for_1 = df[df[category_column] == 1][column].isnull().sum()

            # Dodanie wyników do słownika
            null_counts_dict[column] = {'nulls_for_0': nulls_for_0, 'nulls_for_1': nulls_for_1}

    return null_counts_dict

def plot_nulls_by_category(null_counts_dict):
    # Listy do przechowywania danych do wykresu
    columns = []
    nulls_for_0 = []
    nulls_for_1 = []

    # Wypełnienie list danymi
    for column, counts in null_counts_dict.items():
        columns.append(column)
        nulls_for_0.append(counts['nulls_for_0'])
        nulls_for_1.append(counts['nulls_for_1'])

    # Ustawienia wykresu
    bar_width = 0.35
    index = range(len(columns))

    # Tworzenie wykresu
    plt.figure(figsize=(12, 6))

    # Wykres słupkowy dla nulls_for_0
    plt.bar(index, nulls_for_0, bar_width, label='Nulls for 0', color='blue')

    # Wykres słupkowy dla nulls_for_1 (obok poprzedniego)
    plt.bar([i + bar_width for i in index], nulls_for_1, bar_width, label='Nulls for 1', color='green')

    # Dodanie etykiet i tytułu
    plt.xlabel('Kolumny')
    plt.ylabel('Liczba wartości null')
    plt.title('Liczba wartości null dla każdej kolumny w zależności od wartości w kolumnie kategorii')

    # Ustawienia osi X
    plt.xticks([i + bar_width / 2 for i in index], columns, rotation=45, ha='right', fontsize=10)

    # Dodanie legendy
    plt.legend()

    # Dopasowanie układu
    plt.tight_layout()

    # Wyświetlenie wykresu
    plt.show()

# Przykład użycia funkcji
csv_file = 'twoj_plik.csv'  # Zamień na nazwę swojego pliku CSV
category_column = 'wybrana_kolumna'  # Zamień na nazwę kolumny, której wartości są 0 lub 1

# Obliczenie liczby wartości null i podział na kategorie
null_counts = count_nulls_by_category(csv_file, category_column)

# Wyświetlenie wyników na wykresie
plot_nulls_by_category(null_counts)
```
